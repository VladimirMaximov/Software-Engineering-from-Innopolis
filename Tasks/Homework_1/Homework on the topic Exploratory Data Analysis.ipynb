# Как вы кратко охарактеризуете датасет, про что он?

```python
import numpy as np
import pandas as pd
```

```python
dataset_filename = "Google-Playstore.csv"
data_frame = pd.read_csv(dataset_filename)
data_frame
```

```python
data_frame.info()
```

```python
data_frame.isna().sum()
```

```python
critical_columns_numeric = ["Rating", "Maximum Installs"]
critical_columns_categorial = ["Category"]
column_of_interest = "Maximum Installs"
```

```python
data_frame = data_frame.dropna(subset=critical_columns_numeric)
data_frame = data_frame.query("`Maximum Installs` > 10")
data_frame = data_frame.query("Rating > 0")
data_frame
```

Перед нами датасет из 2_312_944 строк. Он представляет собой набор приложений из Google Playstore со всей необходимой информацией о приложении, среди которой важность имеют столбцы рейтинга, количества скачиваний, а также категории приложения.

Для нас представляет интерес приложения, которыми пользовались люди, соответственно необходимо отфильтровать такие, количество установок у которых больше 10 и рейтинг больше 0, в итоге мы получаем 1_228_668 приложений.

# Сколько в среднем скачиваний у приложений в выборке?

```python
data_frame[column_of_interest].mean()
```

```python
from scipy.stats import trim_mean
trim_mean(data_frame[column_of_interest], 0.10)
```

```python
data_frame[column_of_interest].quantile(0.75)
```

```python
data_frame[column_of_interest].quantile(0.50)
```

```python
data_frame[column_of_interest].quantile(0.25)
```

```python
data_frame[column_of_interest].quantile(0.75) - data_frame[column_of_interest].quantile(0.25)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
percentages = [0.05, 0.1, 0.25, 0.5, 0.75, 0.8, 0.95]
percentiles_df = pd.DataFrame(data_frame[[column_of_interest]].quantile(percentages))
percentiles_df.index = [f'{p * 100}%' for p in percentages]
```

```python
ax = (percentiles_df[column_of_interest]/1_000).plot.box(figsize=(13, 14))
ax.set_ylabel('Maximum Installs (тыс.)')

plt.tight_layout()
plt.show()
```

```python
data_frame["Minimum Installs"].value_counts().nlargest(15).plot(kind='bar', figsize=(15,5))
plt.title("Количество скачиваний приложений")
plt.ylabel('Количество приложений')
plt.xlabel('Minimum Installs')
```

В среднем в выборке от 1_000 до 10_000 скачиваний, подавляющая часть приложений не превышает 30_000 скачиваний.

# Как распределены данные по рейтингу приложений?

```python
ax = (data_frame["Rating"]).plot.hist(density=True, bins=range(0,6), xlim=[0,5], figsize=(14, 4))
(data_frame["Rating"]).plot.density(ax=ax)
ax.set_xlabel('Распределение данных по рейтингу')

plt.tight_layout()
plt.show()
```

Данные по рейтингу приложений распределены довольно плотно. При
этом оценка распределения показывает, что огромная доля приложений кучкуется ассимметрично вокруг верхней границы 
допустимых значений для оценки приложения (т.е. 4-5). Стоит учесть, что при этом приложения, с оценкой меньше 1 не рассматривались ввиду отсутствия интереса на рассмотрение данных приложений.
